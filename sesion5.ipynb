```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

Variables Categoricas

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: str(int(x)))
```

Diagrama de Barras

```python
fig,ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del a√±o')
ax.set_ylabel('Frecuencia')
```

```python
fig,ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del a√±o')
ax.set_ylabel('Frecuencia')
ax.patches[3].set_facecolor('red')
```

```python
fig,ax = plt.subplots()
ax.pie(y.value_counts(),labels=y.value_counts().index)
ax.set_title('Diagrama de Pie')
```

Variables numericas 

```python
y = df['viento']
len(y)
```

```python
fig, ax = plt.subplots()
ax.hist(y,bins=30) #  'bin' particiones de variables continuas
ax.set_xlabel('viento')
ax.set_ylabel('Frecuency')
plt.axvline(np.mean(y),c='r',linestyle = '--',label = 'promedio')
plt.axvline(np.mean(y)+np.std(y),c='k',linestyle = '--',label='+v1-desviacion')
plt.axvline(np.mean(y)-np.std(y),c='g',linestyle = '--',label='+v1+desviacion')
ax.legend()


```
